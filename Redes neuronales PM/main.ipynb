```python
import tensorflow as tf
import numpy as np
```

```python
celsius = np.array ([-20, 100, 50, 45, -42, 13])
kelvin = celsius + 273.5

print(kelvin)
```

```python
capa_oculta_1 = tf.keras.layers.Dense(units=3, input_shape=[1])
capa_oculta_2 = tf.keras.layers.Dense(units=3)
salida = tf.keras.layers.Dense(units=1)
model = tf.keras.Sequential([capa_oculta_1, capa_oculta_2, salida])

model.compile(
    optimizer=tf.keras.optimizers.Adam(0.1),
    loss='mean_squared_error'
)
model.summary()
```

```python
historical = model.fit(celsius, kelvin, epochs=500)
print("Modelo Entrenando")
```

```python
test_values = np.array([-20, 0, 50, 120], dtype=float)

predicciones = model.predict(test_values)

print("Celcius | Kelvin Real | Prediccion NN")
for c, p in zip(test_values, predicciones):
    print(f"{c:7.2f} | {c+273.5:12.2f} | {float(p):.2f}")
```
